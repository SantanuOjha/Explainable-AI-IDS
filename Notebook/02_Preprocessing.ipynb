```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Data/CICIDS2017_combined.csv", low_memory=False)

print("Original shape:", df.shape)
```

```python
df.columns = df.columns.str.strip()
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
print("After cleaning shape:", df.shape)
```

```python
X = df.drop("Label", axis=1)
y = df["Label"]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Number of classes:", len(label_encoder.classes_))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
import gc
from sklearn.preprocessing import StandardScaler

# reduce memory
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

gc.collect()
```

```python
import gc
from sklearn.preprocessing import Normalizer

# reduce memory
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")

norm = Normalizer()

X_train_norm = norm.fit_transform(X_train)
X_test_norm = norm.transform(X_test)

gc.collect()
```

```python
X_train = X_train.astype("float32")
X_test = X_test.astype("float32")

X_train_scaled = X_train_scaled.astype("float32")
X_test_scaled = X_test_scaled.astype("float32")

X_train_norm = X_train_norm.astype("float32")
X_test_norm = X_test_norm.astype("float32")

y_test = y_test.astype("int32")
y_train = y_train.astype("int32")


print("Data types after conversion:")
print("X_train:", type(X_train))
print("X_test:", type(X_test))
print("X_train_scaled:", type(X_train_scaled))
print("X_test_scaled:", type(X_test_scaled))
print("X_train_norm:", type(X_train_norm))
print("X_test_norm:", type(X_test_norm))
print("y_train:", type(y_train))
print("y_test:", type(y_test))
```

```python
pd.DataFrame(X_train).to_csv("../Data/Sampled_Data/X_train.csv", index=False)
pd.DataFrame(X_test).to_csv("../Data/Sampled_Data/X_test.csv", index=False)

pd.DataFrame(X_train_scaled).to_csv("../Data/Sampled_Data/X_train_scaled.csv", index=False)
pd.DataFrame(X_test_scaled).to_csv("../Data/Sampled_Data/X_test_scaled.csv", index=False)
pd.DataFrame(X_train_norm).to_csv("../Data/Sampled_Data/X_train_norm.csv", index=False)
pd.DataFrame(X_test_norm).to_csv("../Data/Sampled_Data/X_test_norm.csv", index=False)
pd.DataFrame(y_train).to_csv("../Data/Sampled_Data/y_train.csv", index=False)
pd.DataFrame(y_test).to_csv("../Data/Sampled_Data/y_test.csv", index=False)
```

```python
np.save("../Data/X_train.npy", X_train)
np.save("../Data/X_test.npy", X_test)
np.save("../Data/X_train_scaled.npy", X_train_scaled)
np.save("../Data/X_test_scaled.npy", X_test_scaled)
np.save("../Data/X_train_norm.npy", X_train_norm)
np.save("../Data/X_test_norm.npy", X_test_norm)
np.save("../Data/y_train.npy", y_train)
np.save("../Data/y_test.npy", y_test)
```
